```python
# filling the null values in the capacity column ov df_agg_bookings with median values
df_agg_bookings['capacity'] = df_agg_bookings['capacity'].fillna(
    df_agg_bookings['capacity'].median()
)
```

```python
df_agg_bookings.shape
```

```python
df_agg_bookings = df_agg_bookings[df_agg_bookings.successful_bookings<=df_agg_bookings.capacity]
df_agg_bookings.shape
```

```python
df_bookings.ratings_given.isnull().sum()
```

```python
median_rating = df_bookings['ratings_given'].median()
df_bookings['ratings_given'] = df_bookings['ratings_given'].fillna(median_rating)
```

## -=> Data Transformation

```python
df_agg_bookings.head()
```

```python
df_agg_bookings["occ_pct"] = df_agg_bookings.successful_bookings / df_agg_bookings.capacity
```

```python
df_agg_bookings.head()
```

```python
df_agg_bookings['occ_pct'] = df_agg_bookings['occ_pct'].apply(lambda x: round(x*100,2))
df_agg_bookings.head()
```
